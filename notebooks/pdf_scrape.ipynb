```python
def process_2018_2021(file, page_num, year):
    df = read_pdf(file, pages = page_num)
    df = df[0].drop(columns = ['Unnamed: 0'])

    second_half_split = df['# of Reports RankState 100K Population'].str.split(" ", 1, expand=True)#.rename({0: '# of Reports', 1: 'Rank State'}, axis = 1)
    df['num reports'] = second_half_split[0]

    second_half_split[['Reports per', 'Unnamed: 1']] = df[['Reports per', 'Unnamed: 1']]
    second_half_split = second_half_split.drop(columns = [0])

    #df_2021['# of Reports'] = second_half_split[0]
    df = df.drop(columns = ['# of Reports RankState 100K Population', 'Reports per', 'Unnamed: 1'])
    second_half_split.columns = df.columns
    df = df.append(second_half_split)
    
    df.columns = ['state', 'id_theft_per_capita', 'id_theft_raw']
    df['year'] = year
    return df
```

```python
df_2021 = process_2018_2021('../consumer sentinel reports/csn_cy-2021_data_book.pdf', '22', 2021)
```

### 2019-2020

```python
df_2020 = read_pdf('../consumer sentinel reports/csn_cy-2019_data_book.pdf', pages = '22')
df_2020 = df_2020[0]
df_2020.columns = ['rankstate', 'reports per 100k', 'skip', 'num reports']
df_2020 = df_2020.drop([0])

df_2020['rankstate'] = df_2020['rankstate'].str.replace(',', '')

split = df_2020['rankstate'].str.split('(\d+)', expand = True)
split = split.drop(columns = [0, 1, 4, 6, 7])
split[8][25] = 'District of Columbia'
split[8][26] = 'Puerto Rico'
split[['reports per 100k', 'num reports']] = df_2020[['reports per 100k', 'num reports']]

rows_to_append = split[[8, 'reports per 100k', 'num reports']]
split = split.drop(columns = [8, 'reports per 100k', 'num reports'])
rows_to_append.columns = split.columns
split = split.append(rows_to_append)
split.columns = ['state', 'id_theft_per_capita', 'id_theft_raw']
```

```python
def process_2019_to_2020(file, page_num, year, opts = None):
    df_2020 = read_pdf(file, pages = page_num, pandas_options = opts)
    df_2020 = df_2020[0]
    df_2020.columns = ['rankstate', 'reports per 100k', 'skip', 'num reports']
    df_2020 = df_2020.drop([0])

    df_2020['rankstate'] = df_2020['rankstate'].str.replace(',', '')

    split = df_2020['rankstate'].str.split('(\d+)', expand = True)
    split = split.drop(columns = [0, 1, 4, 6, 7])
    split[8][25] = 'District of Columbia'
    split[8][26] = 'Puerto Rico'
    split[['reports per 100k', 'num reports']] = df_2020[['reports per 100k', 'num reports']]

    rows_to_append = split[[8, 'reports per 100k', 'num reports']]
    split = split.drop(columns = [8, 'reports per 100k', 'num reports'])
    rows_to_append.columns = split.columns
    split = split.append(rows_to_append)
    split.columns = ['state', 'id_theft_per_capita', 'id_theft_raw']
    split['year'] = year
    return split
```

```python
df_2020 = process_2019_to_2020('../consumer sentinel reports/csn_cy-2020_data_book.pdf', '22', 2020)
df_2019 = process_2019_to_2020('../consumer sentinel reports/csn_cy-2019_data_book.pdf', '22', 2019)
```

```python
def process_2017(file, page_num, year, opts = None):
    df_2017 = read_pdf(file, pages = page_num, pandas_options = {'header': 1})
    df_2017 = df_2017[0].drop([0])

    split = df_2017['Reports per 100K Population'].str.split(' ', expand = True)
    df_2017 = df_2017.drop(columns = ['Reports per 100K Population'])
    df_2017[[0, 1]] = split
    split = df_2017[['Rank State.1', 0, 1]]

    df_2017 = df_2017.drop(columns = split.columns)
    split.columns = df_2017.columns
    df_2017 = df_2017.append(split)
    df_2017.columns = ['state', 'id_theft_per_capita', 'id_theft_raw']
    df_2017['year'] = year
    
    return df_2017
```

```python
df_2017 = process_2017('../consumer sentinel reports/csn_cy-2017_data_book.pdf', '23', year = 2017, opts =  {'header': 1})
```

```python
def convert_to_numeric(x):
    if isinstance(x, str):
        new_str = x.replace(',', '')
        new_str = ''.join(new_str.split())
        new_str = float(new_str)
        return new_str
    else:
        new_str = str(x)
        new_str = new_str.replace(',', '')
        new_str = ''.join(new_str.split())
        new_str = float(new_str)
        return new_str
```

```python
def process_df(df):
    df['state'] = df['state'].str.replace('\d+', '')
    df['id_theft_per_capita'] = df['id_theft_per_capita'].apply(lambda x: convert_to_numeric(x))
    df['id_theft_raw'] = df["id_theft_raw"].apply(lambda x: convert_to_numeric(x))
    return df
```
